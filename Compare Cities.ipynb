```python
import pandas as pd
import Levenshtein
from tqdm import tqdm
```

```python
normalized_cities_df = pd.read_csv("data/normalized_cities - Albert School - normalized_cities.csv")
raw_cities_df = pd.read_csv("data/raw_cities - Albert School - raw_cities.csv")
```

```python
normalized_cities_df
```

```python
cities_df = pd.DataFrame()
cities_df["raw_city"] = raw_cities_df["city"]
cities_df["normalized_city"] = normalized_cities_df["city"]
cities_df["closest_city"] = pd.NA
cities_df["distance"] = pd.NA
cities_df.head()
```

```python
cities_df["raw_city"] = cities_df["raw_city"].fillna("")
```

```python
def closest_match(row, col2):
    closest = float('inf')
    closest_match = None
    distance = float('inf')
    for cell in col2:
        distance = Levenshtein.distance(row, cell)
        if distance < closest:
            closest = distance
            closest_match = cell
    return closest_match, distance

# Extract the two columns we want to compare
col1 = cities_df["raw_city"].to_list()
col2 = cities_df["normalized_city"].to_list()

# Create a new list to store the reorganized rows
reorganized_rows = []

# Iterate through each row in the input file
for index, row in tqdm(cities_df.iterrows()):
    # Extract the cell in the first column
    cell = row["raw_city"]
    if len(str(cell)) == 0:
        continue
    # Find the closest match in the second column and the Levenshtein distance
    closest, distance = closest_match(cell, col2)
    # Replace the cell in the second column with the closest match
    row["normalized_city"] = closest
    # Insert the Levenshtein distance as a new column
    row["distance"] = distance
    # Add the modified row to the list of reorganized rows
    reorganized_rows.append([cell, closest, distance])
```

```python
cities_df
```

```python
import csv
# Write the reorganized rows to a new CSV file
with open('data/cities_output.csv', 'w') as output_file:
    writer = csv.writer(output_file)
    writer.writerows(reorganized_rows)
```
